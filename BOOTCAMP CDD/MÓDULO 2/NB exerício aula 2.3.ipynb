```python
from pyspark.sql import SparkSession
from random import randint
import sys
from operator import add
```

```python
count = 2 * 1000 * 1000

for _ in range(count):
    value = randint(0, 10)
    if value != 5:
        print(value, end=" ")
    else:
        print(value)
```

```python
count = 2 * 1000 * 1000
numbers = []

for _ in range(count):
    value = randint(0, 10)
    if value != 5:
        numbers.append(str(value) + " ")
    else:
        numbers.append(str(value) + "\n")

numbers_final = ''.join(numbers)

with open("numbers.txt", "w") as arquivo:
    arquivo.write(numbers_final)
```

```python
filename = "C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\POS XP\\BOOTCAMP CDD\\MÓDULO 2\\numbers.txt"

spark = SparkSession\
        .builder\
        .appName("PythonNumberCount")\
        .master("local[*]")\
        .getOrCreate()
        
linesRdd = spark.read.text(filename).rdd.map(lambda r: r[0])
print("Number of partitions: {}".format(linesRdd.getNumPartitions()))

counts = linesRdd.flatMap(lambda x: x.split(' '))\
                    .map(lambda x: (x, 1))\
                    .reduceByKey(add)

path = "C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\POS XP\\BOOTCAMP CDD\\MÓDULO 2\\"

output = counts.take(10)
for (number, count) in output:
    print("%s: %i" % (number, count))
```
