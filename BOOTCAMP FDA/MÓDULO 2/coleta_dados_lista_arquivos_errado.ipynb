#importando bibliotecas necessárias

```python
import pandas as pd
import os
```

#criando caminho onde iremos armazenar os dados

```python
localarq = ('/content/drive/MyDrive/Colab Notebooks/PÓS CIENT DADOS - XP/MÓDULO 2/datasets/')
```

```python
lista_df = []
lista_caminho = os.listdir(localarq)
```

```python
lista_caminho
```

```python
lista_df = []
for arquivo in lista_caminho:
  local_arquivo = os.path.join(localarq, arquivo)
  df = pd.read_csv(local_arquivo, sep = '|', encoding = 'latin-1')
  lista_df.append(df)
df_dados = pd.concat(lista_df)
```

```python
df_dados[:10]
```

#exportando

```python
df_dados.to_csv('/content/drive/MyDrive/Colab Notebooks/PÓS CIENT DADOS - XP/MÓDULO 2/merged.csv', sep = ';', encoding = "latin-1", index=False)
```
