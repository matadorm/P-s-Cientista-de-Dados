# '2. Qual vacina é a mais utilizada no mundo?


```python
vacinas_por_pais = dataframe[['country','date','vaccines']]
vacinas_por_pais
```

```python
#agora precisamos agrupar por país
vacinas_por_pais.groupby('vaccines')['country'].nunique().sort_values(ascending=False)
```

```python
vacinas_pais = dataframe.groupby('country')['vaccines'].unique()
vacinas_pais = vacinas_pais.to_frame().reset_index()
vacinas_pais
```

# '3. Qual país possui o maior número de vacinados por dia?

```python
vacinados_dia = dataframe.groupby('country')['daily_vaccinations'].max().sort_values(ascending=False).head()
vacinados_dia = vacinados_dia.to_frame().reset_index()
vacinados_dia
```

```python
linha = vacinados_dia['daily_vaccinations'].idxmax()
linha
```

```python
print(f'O país com maior número de vacinados por dia é {vacinados_dia.iloc[linha,0]} com {vacinados_dia.iloc[linha,1]} vacinas.')
```

# '4. Qual a média de vacinação por dia por país?

```python
media_pais = dataframe.groupby('country')['daily_vaccinations'].mean().sort_values(ascending=False)
media_pais = media_pais.to_frame().reset_index()
media_pais
```

#criando a nova base de dados

##alterando formatação das bases

```python
media_pais = media_pais.rename(columns = {'country':'pais', 'daily_vaccinations':'media_vacinacao'})
media_pais
```

```python
vacinas_pais = vacinas_pais.rename(columns = {'country':'pais', 'vaccines':'vacina'})
vacinas_pais
```

```python
imunizados = imunizados.rename(columns = {'country':'pais', 'total_vaccinations':'imunizados'})
imunizados
```

##exportando
Aqui vamos juntar e exportar
Para juntar, é importante que as colunas usadas para merge estejam nomedas igualmente

```python
dados_tratados = pd.merge(imunizados, media_pais, how='inner')
dados_tratados
```

```python
dados_tratados = pd.merge(dados_tratados, vacinas_pais)
dados_tratados
```
