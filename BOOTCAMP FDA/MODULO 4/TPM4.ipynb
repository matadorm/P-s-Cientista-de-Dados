```python
import pandas as pd

# Carregar dataset
url = 'https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat'
```

```python
#ler dataset
airports = pd.read_csv(url, header=None)
airports
```

```python
#renomear colunas
airports = airports.rename(columns={0:'Airport ID', 1:'Name', 2:'City', 3:'Country', 4:'IATA', 5: 'ICAO', 6:'Latitude', 7:'Longitude', 8:'Altitude', 9:'Timezone', 10:'DST', 11:'Tz database time zone', 12:'Type', 13:'Source'})
airports
```
